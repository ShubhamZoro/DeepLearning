```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
# Load the trained YOLOv8 model


# Load the video
video_path = "/content/drive/MyDrive/Human_detection/human detection dataset/video/ABA Therapy - Social Engagement.mp4"
cap = cv2.VideoCapture(video_path)

# Get the video frame width, height, and FPS to save the output video
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create VideoWriter object to save the video
output_path = "output_video.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Initialize a dictionary to store the unique IDs of tracked persons
tracked_objects = {}
model = YOLO("/content/runs/detect/HumanDetection/weights/best.pt")
# Process the video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Predict using YOLOv8 on the current frame
    results = model.predict(frame)

    # Iterate through detections and assign IDs
    for result in results:
        boxes = result.boxes  # Get the bounding boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Get bounding box coordinates
            confidence = box.conf[0].item()  # Get confidence level
            class_id = int(box.cls[0])  # Get class ID

            # Assuming 0 is Child and 1 is Therapist in the dataset
            label = "Child" if class_id == 0 else "Therapist"

            # Assign a unique ID and track the person
            if confidence > 0.5:
                if class_id not in tracked_objects:
                    tracked_objects[class_id] = len(tracked_objects) + 1

                person_id = tracked_objects[class_id]
                cv2.putText(frame, f"{label} {person_id}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

    # Write the frame with detections to the output video
    out.write(frame)

    # Display the frame for debugging (optional)
    cv2_imshow(frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release everything once job is finished
cap.release()
out.release()
cv2.destroyAllWindows()

print("Detection and tracking completed. Output video saved.")
```
